##### Buffering HUC8 watersheds by 5km

```python
# load packages
import geopandas as gpd
from pathlib import Path
from zipfile import ZipFile
```

```python
# path to folders
input_zip = Path("huc8_layer.zip")
output_dir = Path("covariates"); output_dir.mkdir(exist_ok=True)
```

```python
# find the shapefile in the uploaded zipfile and read it
with ZipFile(input_zip) as z:
    shp_path = [f for f in z.namelist() if f.endswith(".shp")][0]

gdf = gpd.read_file(f"zip://{input_zip}!{shp_path}")
```

```python
# reproject to UTM Zone 15N
if gdf.crs.is_geographic:
    gdf = gdf.to_crs("EPSG:32615")

# make the 5km buffer
gdf["geometry"] = gdf.buffer(5000)

# save to geopackage
gdf.to_file(output_dir / "huc8_5km_buffer.gpkg", driver="GPKG")

# know that it worked
print("buffered watershed complete")
```
